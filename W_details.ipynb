## Calculating Age

```python
def Age(born):
    today = date.today()
    try: 
        birthday = born.replace(year=today.year)
    except ValueError: # raised when birth date is February 29 and the current year is not a leap year
        birthday = born.replace(year=today.year, month=born.month+1, day=1)
    if birthday > today:
        return today.year - born.year - 1
    else:
        return today.year - born.year
```

```python
df.loc[:,'Age'] = df.DOB.apply(Age)
df
```

## Setting Age Ranges

```python
def age_catg(age):
    if age < 30: r = 29
    elif age < 45 : r = 44
    elif age < 65 : r = 64
    else: r = 100
    return r
```

```python
df['age_range'] = df['Age'].apply(age_catg)
df
```

```python
fig,axes = plt.subplots(2,2,figsize=(15,15))
sns.histplot(data = df, x = 'Age',binwidth = 10,color = "green",kde=True,ax=axes[0,0])
sns.histplot(data = df, x = 'age_range',binwidth = 50,color = "blue",kde=True,ax=axes[1,0])
sns.histplot(data = df, x = 'Cname',binwidth = 10,color = "red",kde=True,ax=axes[0,1])
sns.histplot(data = df, x = 'State',binwidth = 1000,color = "purple",kde=True,ax=axes[1,1])
```

## Bar plot between City name and age_range

```python
x = df.Cname
y = df.age_range
```

```python
df.dtypes
```

```python
plt.figure(figsize = (15, 8))
plt.xticks(rotation=0, ha='center')

barplt=sns.barplot(x=x,y=y)

for p in barplt.patches:
    barplt.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 4, p.get_height()), 
                   ha = 'left', va = 'center', 
                   xytext = (0, 10), 
                   textcoords = 'offset points')

plt.title('City wise Age Range')
plt.xlabel('City')
plt.ylabel('Age Range')
```
