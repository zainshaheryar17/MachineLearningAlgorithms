```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("E://DATA SETS//emp.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df = df.convert_dtypes()
df.dtypes
```

```python
df['hiredate'] = pd.to_datetime(df['hiredate'])
df.dtypes
```

```python
df.head()
```

```python
df['Quarter'] = df['hiredate'].apply(lambda x: x.quarter)
df['Year'] = df['hiredate'].apply(lambda x: x.year)
df['Month'] = df['hiredate'].apply(lambda x: x.month)
df['Week'] = df['hiredate'].apply(lambda x: x.week)
df['Day'] = df['hiredate'].apply(lambda x: x.day)

df
```

```python
df['Quarter'] = df.Quarter.astype('str')
```

```python
df.dtypes
```

```python
qrtr = {'1':'Q1', '2': 'Q2', '3':'Q3', '4':'Q4'}
df['Quarter'] = df['Quarter'].map(qrtr)
df
```
