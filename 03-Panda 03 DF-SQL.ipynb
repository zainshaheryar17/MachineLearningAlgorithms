```python
e = emp
d = pd.read_csv("E:\DATA SETS\dept.csv")
print(e.columns)
print(d.columns)
```

```python
pd.merge(e, d) # JOINS all same column(s)name
```

```python
pd.merge(e, d, on='deptno', how='inner') # JOIN specific same column(s) name. Joining type
```

```python
pd.merge(e, d, left_on='deptno', right_on = 'deptno', how='inner')
```

```python
emp.groupby(['deptno', 'job']).agg(TotalEmp = ('empno', 'count'), \
       AvgSal = ('sal','mean'), Sumsal = ('sal', 'sum')).reset_index()
#SELECT deptno, job, COUNT(empno) AS TotalEmp, AVG(sal) AS Avgsal, SUM(sal) as SumSal
#from emp
#GROUP BY deptno, job
```

```python
emp.groupby(['deptno', 'job']).agg(TotalEmp = ('empno', 'count'), \
       AvgSal = ('sal','mean'), Sumsal = ('sal', 'sum'))
# Removal of resetindex function means that group by columns will be treated as index. If we use resetindex function
# it will create its own index.
```

```python
pd.merge(e, d, on='deptno', how='right') # 'left'; 'outer'
```

```python
pd.merge(e, d, on='deptno', how='outer') # 'outer'
```

```python
pd.merge(e, d, on='deptno', how='outer', indicator = True) 
```
