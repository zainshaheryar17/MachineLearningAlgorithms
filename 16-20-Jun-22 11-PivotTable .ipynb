```python
pd.pivot_table(df, index=["Manager", "Status"], columns=["Product"], aggfunc={"Quantity":'sum', "Amount":'mean'},\
               fill_value=0)

#here, we wanted to use sum on quantity and mean on amount. Prviosuly, we were using aggfunc on all stuff in values.
#now we are particularly telling that we need sum of qty and mean of amount.
```

```python
table = pd.pivot_table(df, index=["Manager", "Status"], columns=["Product"], values=["Quantity", "Amount"],\
               aggfunc={"Quantity":'sum', "Amount":['sum', 'mean']},\
               fill_value=0)
table

#here, we want to see sum of qty and sum and mean of amount. we are usning multiple aggregate function here.
```
