```python
#lets use Seaborn to draw a histogram
```

```python
plt = sns.distplot(df['salary'])
plt.set_xlabel("Salary, fontsize = 15")

#ylabel is density by default
```

```python
#Use regular plot function to display a barplot

plt = df.groupby(['rank'])['salary'].count().plot(kind='bar')
plt.set(title='Bar Plot')
plt.set( xlabel = "Rank", ylabel = "Count")
for p in plt.patches:
    plt.annotate(format(p.get_height()),
            (p.get_x() + p.get_width()/2., p.get_height()),
            ha= 'center' , va = 'center', xytext = (0,9),
            textcoords = 'offset points')
```

```python
# use seaborn to display barplot
sns.set_style("whitegrid")
plt = sns.barplot(x='rank', y = 'salary', data=df, estimator=sum)
plt.set(title='SNS Bar Plot')
plt.set( xlabel = "Rank", ylabel = "Salary")
for p in plt.patches:
        plt.annotate(format(p.get_height()),
                    (p.get_x() + p.get_width()/2., p.get_height()),
                    ha= 'center' , va = 'center', xytext = (0,9),
                    textcoords = 'offset points')
```

```python
#split into 2 groups
chrt = sns.barplot(x='rank', y = 'salary', hue='gender', data=df, estimator=sum)
chrt.set(title='SNS Bar Plot')
chrt.set( xlabel = "Rank", ylabel = "Salary")
for p in plt.patches:
    plt.annotate(format(p.get_height()),
                 (p.get_x() + p.get_width()/2., p.get_height()),
                  ha= 'center' , va = 'center', xytext = (0,9),
                  textcoords = 'offset points')
```
